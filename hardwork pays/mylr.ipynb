```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
inp=pd.read_csv("E:\practise\Linear_X_Train.csv")
out=pd.read_csv("E:\practise\Linear_Y_Train.csv")

x=np.array(inp)
y=np.array(out)


```

```python
plt.figure()
plt.scatter(x,y)
plt.style.use("seaborn")
plt.show()
```

```python
class lin_regre():
    def __init__(self,x,y):
        self.x=x
        self.y=y
        self.coef=np.random.uniform(low=-1,high=1)
        self.bias=np.random.random()
        
    def hypo(self,x):
        return self.bias+self.coef*x
    
    def grad(self):
        grad0=0
        grad1=0
        for x,y in zip(self.x,self.y):
            grad0+=(self.hypo(x)-y)
            grad1+=(self.hypo(x)-y)*x
        return grad0/self.x.shape[0] , grad1/self.x.shape[0]
    
    def updatew(self,lr=0.1):
        grad0_ans,grad1_ans=self.grad()
        self.coef=self.coef-lr*grad1_ans
        self.bias=self.bias-lr*grad0_ans
    def predict(self,x):
        yhat=[]
        for i in x:
            yhat.append(self.hypo(i))
        return yhat
    def loses(self):
        
        loss=[]
        for x,y in zip(self.x,self.y):
            loss.append((self.hypo(x)-y)**2)
        loss=np.array(loss)
        return loss.sum()/self.x.shape[0]*2
    def run(self,a=10):
        loss=[]
        for i in range(a):
            self.updatew()
            ith_loss=self.loses()
            print("loss at {}th iteratio is {}  ".format(i,ith_loss))
            self.plot()
            loss.append(ith_loss)
        loss=np.array(loss)
        plt.figure()
        plt.style.use("seaborn")
        plt.plot(loss)
        return self.loses()
    def plot(self):
        plt.figure()
        plt.style.use("seaborn")
        plt.scatter(self.x,self.y)
        pred=self.predict(self.x)
        pred=np.array(pred)
        pred.shape
        plt.plot(self.x,pred,color="red")
        plt.show()        

    
    
    
    
    
```

```python
l=lin_regre(x,y)


```

```python
loss=[]
loss.append(l.loses())
print(loss)
```

```python
l.plot()
```

```python
l.run(50)
```
